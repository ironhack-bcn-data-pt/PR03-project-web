```python
import pandas as pd
import numpy as np

df2 = pd.read_csv('justETF.csv')
```

```python
df2=df2.replace(['-'], np.nan )
```

```python
df2.keys()
```

```python
df2['Fund size [M]'] = df2['Fund size [M]'].str.replace(',','').astype(float)
```

```python
for column in df2.columns[5:]:
   df2[column] = df2[column].str.replace('%','').astype(float)
```

# Visado

```python
#define the url
url_diario = 'https://www.eldiario.es/viajes/guias_y_consejos/paises-turisticos-piden-visado-espanoles_1_2811889.html'
```

```python
#request the url
response_diario = requests.get(url_diario, headers=headers)
```

```python
#cheking if it works
response_diario.ok
```

```python
#get the contents 
diario = BeautifulSoup(response_diario.content, "html.parser")
```

```python
#find the parameters that we want
articulo = diario.findAll('p', attrs={'class':'article-text'})
```

```python
diario.findAll('p', attrs={'class':'article-text'})[10].find('strong').text
```

```python
#getting the list of countries that requires visa
visado=[]
for el in articulo:
    if el.find('strong')!=None:
        visado.append(el.find('strong').text)
```

```python
#link of countries translated
url_trad ='https://www.101languages.net/spanish/country-names-spanish/'
```

```python
visado
```

```python
response_trad = requests.get(url_trad, headers=headers)
```

```python
response_trad.reason
```

```python
response_trad.ok
```

```python
trad = BeautifulSoup(response_trad.content, "html.parser")
```

```python
trad.findAll('tr')
```

```python
#finding the countries translated
tr = trad.findAll('tr')
```

```python
tr[1].findAll('td')[0]
print(len(tr))
```

```python
traducciones=[]
for i in range(1,len(tr)):
    traducciones_dicc={
    'english':tr[i].findAll('td')[0].text.strip(),
    'spanish':tr[i].findAll('td')[1].text.strip()
    }
    traducciones.append(traducciones_dicc)
```

```python
traducciones[0]['spanish']
```

```python
countries = []
for elem in visado:
    for pais in traducciones:
        if elem in pais['spanish']:
            countries.append(pais['english'])
        
```

```python
#we will get all ratings from the differents links of each airport
ratings=[]
for el in SPAIN['airport']:
    url =f'https://api.flightradar24.com/common/v1/airport.json?code={el}'
    resp_el = requests.get(url, headers=headers)
    rate = resp_el.json()['result']['response']['airport']['pluginData']['flightdiary']['ratings']['avg']
    rate_dicc={
        'airport':el,
        'rate':rate
    }
    ratings.append(rate_dicc)
```

```python
ratings_df = pd.DataFrame(ratings)
```

```python
Aeropuertos = pd.merge(SPAIN, ratings_df, how='left', on='airport')
```

```python
Aeropuertos.head()
```

```python
#best rated airport in spain
Aeropuertos[Aeropuertos['rate']==Aeropuertos['rate'].max()]
```

```python
#worst rated airport in spain
Aeropuertos[Aeropuertos['rate']==Aeropuertos['rate'].min()]
```

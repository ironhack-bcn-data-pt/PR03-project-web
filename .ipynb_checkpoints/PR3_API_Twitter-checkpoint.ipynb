##### Para obtener el resto de los usuarios hay otra manera de pedir la info, usando el Cursor

Tweepy tiene una clase cursor puedes obtener los items que quieres sin preocuparte por la paginaci√≥n:

```python



#A la clase (tweepy.Cursor) le pasas el m√©todo -sin llamarlo- (api.followers) 
#Le pasas los par√°metros del m√©todo al que llamo (screen_name = "myHRfuture")
#Le dices a Cursos cu√°ntos items quieres obtener (.items(el n√∫mero que quieres))
#Iteras el cursor (for user in)
#Dentro del for recuperas el print de los usuarios del json

for user in tweepy.Cursor(api.followers, screen_name = "myHRfuture").items(50):
    print(json.dumps(user._json, indent = 2))
```

#### C√≥mo sacar los followees de una cuenta de Twitter:

```python
#Vamos a usar el m√©todo cursor otra vez para sacar los followees, o sea a qui√©n sigue myHRfuture
#s√≥lo hay que cambiar de api.followers a api.friends

for user in tweepy.Cursor(api.friends, screen_name = "myHRfuture").items(50):
    print(json.dumps(user._json, indent = 2))
```

#### C√≥mo obtener el timeline de un usuario (todos los tweets que ha hecho un usuario):

```python
#Vamos a usar el cursor otra vez para no tener l√≠mites de paginaci√≥n
#Cambias el m√©todo (api.user_timeline)
#A√±ades otro par√°metro, tweet_mode, que determina la longitud del tweet que quiero obtener. Le pones extended para obtener todo el contenido de los tweets de m√°s de 140 caract√©res
#Tambi√©n he cambiado en el for "user" por "tweet" para que se adapte a nuestro request

for tweet in tweepy.Cursor(api.user_timeline, screen_name = "myHRfuture", tweet_mode = "extended").items(10):
    print(json.dumps(tweet._json, indent = 2))
```

```python
#El texto del tweet est√° en la clave "full_text"
#Te devuelve MUCHA informaci√≥n de cada tweet, como por ejemplo el usuario que le di√≥ like o lo retwite√≥
```

```python
#Aqu√≠ el m√©todo es api.search
#Como par√°metro, si lo que buscas es texto, usas q = "palabra clave"

#Para guardar el contenido del for en una variable, guardo todas las iteraciones del for en un string 

jt = ""
for tweet in tweepy.Cursor(api.search, q = "People Analytics", tweet_mode = "extended").items(100):
    jt += json.dumps(tweet._json, indent = 2)
```

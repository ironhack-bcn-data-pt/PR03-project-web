# THE MOVIE DATA BASE API

### Importing libraries

```python
import requests
import json
import pandas as pd
```

### Finding id for Billy Wilder

```python
url = 'https://api.themoviedb.org/3/search/person?api_key={}&query=billy+wilder'
response = requests.get(url)
data = json.loads(response.text)
print(data)
```

> The id for Billy Wilder is 3146

### Get more info with function

```python
query = {'api_key':''}
```

```python
api_url = 'https://api.themoviedb.org/3/person/'
```

```python
def get_person(person_id):
    query = {'api_key':''}
    api_url = 'https://api.themoviedb.org/3/person/'
    return requests.get(f'{api_url}{person_id}', params = query).json()
```

```python
get_person(3146)
```

### Getting all the filmography

```python
credits = 'https://api.themoviedb.org/3/person/3146?api_key={}&append_to_response=credits'
response = requests.get(credits)
movies = json.loads(response.text)
print(movies)
```

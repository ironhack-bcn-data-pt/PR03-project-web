```python
driver.get('http://continente.pt')
driver.set_window_size(1024, 768)
driver.implicitly_wait(2)
```

```python
cookies_pop_up = driver.find_element(by='id', 
                                 value='CybotCookiebotDialogBodyLevelButtonLevelOptinAllowAll')
time.sleep(2)
cookies_pop_up.click()
```

```python
def clickMenu():
    categories_list = driver.find_element(
    by='css selector', value='.icon-menu')
    categories_list.click()
```

```python
## Go back /html/body/div[2]/header/nav/div/div[2]/div/nav/div[3]/ul/li[1]/ul/li[1]/div[1]
```

```python
#Navigate through the different categories
def NavigateCat1():
    categories = driver.find_elements(by='css selector', value ='''ul.dropdown-toggle-menu:nth-child(1) > li:nth-child(1)''')
    for category in categories:
        try:
            category.click()
            time.sleep(0.5)
            NavigateCat2()
            continue
        except:
            break
```

```python
def NavigateCat2():
    categories2 = driver.find_elements(by='css selector', value ='''.in > li''')
    for category2 in categories2:
        try:
            category2.click()
            time.sleep(0.5)
            NavigateCat3()
        except:
            pass
```

```python
def NavigateCat3():
    categories3 = driver.find_elements(by='css selector', value ='''ul.dropdown-toggle-menu:nth-child(1) > li:nth-child(1) > ul:nth-child(4) > li:nth-child(1) > ul:nth-child(4) > li:nth-child(1)''')
    for category3 in categories3:
        try:
            category3.click()
            time.sleep(0.5)
            ScrapProducts() 
        except:
            pass
```

```python
def Scroll():
    SCROLL_PAUSE_TIME = 0.5

    # Get scroll height
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        # Scroll down to bottom
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Wait to load page
        time.sleep(SCROLL_PAUSE_TIME)

        # Calculate new scroll height and compare with last scroll height
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
```

```python
def ScrapProducts():
    Scroll()
    products = driver.find_elements_by_class_name('product')
    len(products)
    for product in products:
        try:
            product_name = product.find_element_by_class_name('ct-tile--description')
            print(product_name.text)
            product_picture = product.find_element(by='css selector', value ='.ct-tile-image').get_attribute('src')
            print(product_picture)
            product_brand = product.find_element_by_class_name('ct-tile--brand')
            print(product_brand.text)
            product_quantity = product.find_element_by_class_name('ct-tile--quantity')
            print(product_quantity.text)
            product_price = product.find_element_by_class_name('ct-price-formatted')
            print(product_price.text)
            try:
                product_price_discounted = product.find_element_by_class_name('ct-tile--price-dashed')
                print(product_price_discounted.text)
            except:
                pass
            product_price_quantity = product.find_element_by_class_name('ct-price-value')
            print(product_price_quantity.text)  
        except:
            pass    
```

```python
def RunScrapping():
    try:
        clickMenu()
        NavigateCat1()
    except:
        pass
```

```python
RunScrapping()
```

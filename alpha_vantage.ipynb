```python
df_volume = update_df.drop(['Meta Data', 'Time Series (Daily)'], axis=1)
```

```python
df_volume
```

```python
df_volume.Volume = df_volume.Volume.astype(int)

```

```python
df_volume = df_volume.reset_index()
```

```python

df_top_toplot = df_volume.nlargest(10,'Volume').reset_index().sort_values(by="index")
```

```python
fig, ax = plt.subplots()
ax.xaxis.set_major_locator(matplotlib.dates.DayLocator(interval=1))

plt.rcParams["figure.figsize"] = (60, 10)
ax.plot(df_volume["index"], df_volume["Volume"])
fig.autofmt_xdate()
plt.show()
```

```python
df_top_toplot.plot(kind='bar' , x='index', y='Volume', title="Top Volume Days");
```
